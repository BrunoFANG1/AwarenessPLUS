```python
import pandas as pd

# Read input and data files and convert them into a list of documents for clustering
input_dir = './M1_input.txt'
input_file = open(input_dir, 'r')

csv_dir = './polusa/2017_1.csv'
data_file = pd.read_csv(csv_dir)
data_arr = data_file.to_numpy()

docs = [] 
docs.append(input_file.read())
for row in data_arr: 
    docs.append(row[5])
```

```python
from bertopic import BERTopic
from bertopic.vectorizers import ClassTfidfTransformer
from sklearn.datasets import fetch_20newsgroups
import time

# # Data
# docs = fetch_20newsgroups(subset='all',  remove=('headers', 'footers', 'quotes'))['data']

time1 = time.time()

# Load and run model
# topic_model = BERTopic()
ctfidf_model = ClassTfidfTransformer()
topic_model = BERTopic(ctfidf_model=ctfidf_model)
topics, probs = topic_model.fit_transform(docs)

time2 = time.time()
print(time2-time1)
```

```python
# A Pandas Dataframe of all topic infos
topic_model.get_topic_info()
```

```python
# A Pandas Dataframe of all document infos
topic_model.get_document_info(docs)
```

```python
# Convert document info dataframe to np array 
# arr[][0] = documents
# arr[][1] = topics
docs_arr = topic_model.get_document_info(docs).to_numpy()

# Get all documents of a given topic, e.g., topic 0
topic = docs_arr[0][1]
docs_of_topic = []
for row in docs_arr: 
    if row[1] == topic:
        docs_of_topic.append(row[0])
        
print(topic)
print(docs_of_topic[0])
```
